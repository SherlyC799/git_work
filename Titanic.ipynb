```python
d.drop('Cabin',axis=1,inplace=True)
```

```python
d.dropna(inplace=True)
```

```python
sns.heatmap(d.isnull(),yticklabels=False)
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
d['Sex']=l.fit_transform(d['Sex'])
d.head(20)
```

```python
d['Embarked']=l.fit_transform(d['Embarked'])
d.head(20)
```

```python
d['Pclass']=l.fit_transform(d['Pclass'])
d.head(20)
```

```python
d.drop(['PassengerId','Name','Ticket','Pclass'],axis=1, inplace=True)
```

```python
d.head(20)
```

```python
from sklearn.model_selection import train_test_split
x=d.drop(['Survived'],axis=1)
y=d['Survived']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.linear_model import LogisticRegression
s=LogisticRegression()
s.fit(x_train,y_train)
ypred=s.predict(x_test)
from sklearn.metrics import classification_report,accuracy_score
c=classification_report(y_test,ypred)
print('Classification Report:\n', c)
```
