```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the frequency of each unique value in 'Dosages Description'
dosage_counts = df['Dosages Description'].value_counts()

# Plot the bar chart using the index (unique values) as x and the values (frequencies) as height
plt.bar(dosage_counts.index, dosage_counts.values, color='skyblue')
plt.title('Dosages Description')
plt.xlabel('Dosages Description')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python

ds = df.query('`Dosages Description` == "Capsule"')
# Group by 'Generic Name and Strength' and get the count of each unique drug.
drug_counts = ds['Generic Name and Strength'].value_counts()
```

```python
da=drug_counts.sort_values( ascending=False).head()
da
```

```python
plt.figure(figsize=(15,8))
my=[0.2,0,0,0,0]
# Change 'color' to 'colors' to specify the color of the slices.
plt.pie(da.values, labels=da.index ,explode=my, shadow=True )
plt.title('Mostly Used Drugs (Capsules)')
plt.xlabel('Drug names')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend()
plt.show() # Add this line to display the pie chart
```

```python
#Most used drug by the companies
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 5 manufacturers
top_manufacturers = df['Name of the Manufacturer'].value_counts().head().index

# Filter the original DataFrame for these manufacturers
filtered_df = df[df['Name of the Manufacturer'].isin(top_manufacturers)]

# Now calculate value counts of 'Generic Name and Strength' within this filtered DataFrame
de = filtered_df['Generic Name and Strength'].value_counts().head()

de
```

```python
plt.figure(figsize=(15,8))
plt.pie(de.values,labels=de.index, shadow=True, autopct='%1.1f%%')
plt.title('Top 5 Drugs Used by Manufacturers')
plt.xlabel('Drug names')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
